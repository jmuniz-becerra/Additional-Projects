```python
import ast
```

```python
# Deconstructing the ratings data to be able to work with it
ast.literal_eval('[1, 2, 3]')
```

```python
# Deconstructing of ratings data
ast.literal_eval(df.ratings[0])
```

```python
# Creation of function to deconstruct ratings data
def str_to_list(ratings_str):
    return ast.literal_eval(ratings_str)
```

```python
str_to_list(df.ratings[0])
```

```python
df.ratings.apply(ast.literal_eval).head()
```

```python
# Creation of new ratings_list column and incorporation of lambda function
df['ratings_list'] = df.ratings.apply(lambda x: ast.literal_eval(x))
```

```python
# Total number of ratings received by each talk?
```

```python
df.ratings_list[0]
```

```python
# Creation of function to get num ratings for each talk
def get_num_ratings(list_of_dicts):
    num = 0
    for d in list_of_dicts:
        num = num + d['count']
    return num
```

```python
get_num_ratings(df.ratings_list[0])
```

```python
# Creation of num ratings column
df['num_ratings'] = df.ratings_list.apply(get_num_ratings)
```

```python
df
```

```python
# Which occupations delivered the funniest TED talks on average?
```

```python
df.ratings_list.head()
```

```python
# Calculating the number of funny ratings
df.ratings.str.contains('Funny').value_counts()
```

```python
# Creation of function to get funny ratings for talk
def get_funny_ratings(list_of_dicts):
    for d in list_of_dicts:
        if d['name'] == 'Funny':
            return d['count']
```

```python
# Creation of funny ratings column for data
df['funny_ratings'] = df.ratings_list.apply(get_funny_ratings)
```

```python
df.funny_ratings.head()
```

```python
df['funny_rate'] = df.funny_ratings / df.num_ratings
```

```python
# Assessing the most funniest occupations from TED talks
df.sort_values('funny_rate').speaker_occupation.head(20)
```

```python
df.groupby('speaker_occupation').funny_rate.mean().sort_values().tail()
```
